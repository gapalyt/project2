```python
# 시각화 - 도로 분류

# 한글 깨짐
!apt -qq -y install fonts-nanum > /dev/null

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

fontpath = 'C:\Windows\Fonts\맑은 고딕/malgun.ttf'
font = fm.FontProperties(fname=fontpath, size=10)
fm._rebuild()

# 그래프에 retina display 적용
%config InlineBackend.figure_format = 'retina'

# Colab 의 한글 폰트 설정
plt.rc('font', family='NanumBarunGothic') 

# 그래프 크기 고정
plt.rcParams["figure.figsize"] = (12, 9)

# 음수 기호가 깨지는 것을 방지
plt.rcParams['axes.unicode_minus'] = False

# bar 차트
x = df_icing['도로 분류'].value_counts().index.tolist()
y = df_icing['도로 분류'].value_counts().values.tolist()
plt.xlabel('도로 분류')
plt.ylabel('수')
plt.title('도로 분류 비교')
plt.bar(x, y)

plt.show()
```

```python
# barplot 차트
import seaborn as sns

x = df_icing['도로 분류'].value_counts().index.tolist()
y = df_icing['도로 분류'].value_counts().values.tolist()
sns.barplot(x = x, y = y, palette = 'Paired')
plt.xlabel('도로 분류')
plt.ylabel('수')
plt.title('도로 분류 비교')
plt.legend(x)

# 값 표시
for index, value in enumerate(y):
  plt.text(index - 0.1, value + 10, str(value))  

plt.show()
```

```python
# 파이 차트
x = df_icing['도로 분류'].value_counts().index.tolist()
y = df_icing['도로 분류'].value_counts().values.tolist()
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

plt.pie(y, labels = x, autopct = '%.2f%%', startangle = 260, counterclock = False, wedgeprops = wedgeprops)
plt.title('도로 분류 비교')
plt.legend(loc = 'best')

plt.show()
```
