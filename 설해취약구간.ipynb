## 설해취약구간

```python
import requests

url = 'http://data.ex.co.kr/openapi/ctsrRiskMap/snowLeakSect?key=7284801427&type=json&numOfRows=69&pageNo=1'

res = requests.get(url)

datas = res.json()
datas
```

```python
data = datas['snowLeakSecLists'] # 필요한 데이터 추출
data
```

```python
import pandas as pd

dfs = pd.DataFrame(data)
dfs
```

```python
df = dfs.drop(['message', 'code'], axis = 1) # message, code 제거
df.columns = ['노선', '본부', '지사', '시점(km)', '종점(km)', '종단선형', '평면선형', '선정사유'] # 이름 변경

df
```

## csv 파일로 저장

```python
df.to_csv('설해취약구간.csv', encoding='utf-8-sig')
```
