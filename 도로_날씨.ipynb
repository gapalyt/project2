```python
import pandas as pd

df_weather = pd.read_csv('/content/drive/MyDrive/멀티캠퍼스/I/파이널 프로젝트/자료/도로 날씨_수정.csv')
df_weather
```

```python
df_weather.drop(['Unnamed: 0'], axis = 1, inplace = True)
df_weather
```

```python
# 발표일시 datetime으로 변환
df_weather['발표일시'] = df_weather['발표일시'].astype('str')
df_weather['발표일시'] = pd.to_datetime(df_weather['발표일시'], errors='coerce')
df_weather['발표일시'] = pd.to_datetime(df_weather['발표일시'])
```

```python
# 현재기온 nan값 제거 + 인덱스 재배열
df_weather_nan = df_weather[df_weather['현재기온'] == '-999.000000'].index
df_weather = df_weather.drop(df_weather_nan)
df_new_weather1 = df_weather.reset_index(drop = True)

df_new_weather1
```

```python
df_new_weather1['현재기온'].dtype
```

```python
# 현재기온 float로 변환
df_new_weather2 = df_new_weather1['현재기온']
df_new_weather3 = df_new_weather2.apply(pd.to_numeric, errors='coerce')

df_new_weather3
```

```python
df_new_weather1['현재기온'] = df_new_weather3
df_new_weather1
```

```python
# 날짜 정렬
df_new_weather1.sort_values(by=['발표일시'], inplace = True, ascending = True)
df_new_weather1
```

```python
df_new_weather1.isnull() # 결측치 전체 확인
```

```python
df_new_weather4 = df_new_weather1.dropna() # 결측치 제거
df_new_weather4
```

```python
# 날짜 정렬
df_new_weather4.sort_values(by=['발표일시'], inplace = True, ascending = True)
df_new_weather4
```

```python
# 발표일시를 인덱스에 
df_new_weather = df_new_weather4.set_index(keys=['발표일시'], inplace = False, drop = False)
df_new_weather
```

```python
df_name_all = df_new_weather['기상실황지역코드'].value_counts().index.tolist()
df_name_all
```

```python
len(df_name_all)
```
